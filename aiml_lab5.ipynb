```python
#importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#Reading the dataset
df = pd.read_csv('aiml-lab4.csv')
```

```python
# Histograms for each numerical column
df.hist(figsize=(12, 10), bins=30)
plt.tight_layout()
plt.show()
```

```python
# Boxplots for each numerical column
plt.figure(figsize=(12, 10))
for i, column in enumerate(df.select_dtypes(include=['float64', 'int64']).columns):
    plt.subplot(3, 3, i + 1)
    sns.boxplot(y=df[column])
    plt.title(column)
plt.tight_layout()
plt.show()
```

```python
# Pairplot
sns.pairplot(df)
plt.show()
```
